```python
df.drop(['School or Citywide Complaint','Vehicle Type','Taxi Company Borough','Taxi Pick Up Location','Garage Lot Name','Ferry Direction','Ferry Terminal Name','Bridge Highway Name','Bridge Highway Direction','Road Ramp','Bridge Highway Segment','Landmark','Intersection Street 1','Intersection Street 2','School Not Found','School Zip','School Code','School Region','Facility Type'],axis=1,inplace=True)
```

```python
df['Created Date'] = pd.to_datetime(df['Created Date'])
df['Closed Date'] = pd.to_datetime(df['Closed Date'])
```

```python
df.info()
```

```python
df.head()
```

```python
df['Request_Closing_Time']= (df['Closed Date'] - df['Created Date']).dt.total_seconds() / 60.0
```

```python
df.info()
```

```python
df['Request_Closing_Time_days']= (df['Closed Date'] - df['Created Date']).dt.days
```

```python
df.dropna(how='any',inplace= True)
```

```python
df['Request_Closing_Time']= df['Request_Closing_Time'].astype(int)
```
